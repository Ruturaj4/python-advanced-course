```python
row = "||\t\t||\t\t||\t\t||\n||\t{}\t||\t{}\t||\t{}\t||\n||\t\t||\t\t||\t\t||\n==\t\t==\t\t==\t\t==\n"

def board(dic):
    return row.format(dic[7],dic[8],dic[9]) + row.format(dic[4],dic[5],dic[6]) + row.format(dic[1],dic[2],dic[3])
```

```python
def check(dic, a):
    return (dic[1] == dic[2] == dic[3] == a) or (dic[4] == dic[5] == dic[6] == a) or (dic[7] == dic[8] == dic[9] == a) or (dic[1] == dic[2] == dic[3] == a) or (dic[1] == dic[4] == dic[7] == a) or (dic[2] == dic[5] == dic[8] == a) or (dic[3] == dic[6] == dic[9] == a) or (dic[1] == dic[5] == dic[9] == a) or (dic[3] == dic[5] == dic[7] == a)
```

```python
def welcome():
    dic = {7:"", 8:"", 9:"", 4:"", 5:"", 6:"", 1:"", 2:"", 3:"",}
    print("Welcome strangers! Please enter your names.")
    player1 = input("Player1: ")
    player2 = input("Player2: ")
    print("Hey {} and {}!".format(player1, player2))
    print(board(dic))
    x_in = int(input("{} will play first. Enter a no. (1 to 9): ".format(player1)))
    dic[x_in] = "X"
    print(board(dic))
    # Count of x and o
    x = 0
    o = 0
    for v in range(4):
        o_in = int(input(f"{player2}, enter a no.: "))
        dic[o_in] = "O"
        print(board(dic))
        o += 1
        if check(dic, "O"):
            winner = player2
            break
        x_in = int(input("%s, enter a no.: " % player1))
        dic[x_in] = "X"
        print(board(dic))
        x += 1
        if check(dic, "X"):
            winner = player1
            break
    print(board(dic))
    print("{} wins!".format(winner))
```
